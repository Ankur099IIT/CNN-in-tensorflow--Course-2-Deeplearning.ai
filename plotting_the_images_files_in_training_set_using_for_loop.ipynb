```python
import zipfile

zip_ref = zipfile.ZipFile('/content/rps.zip', 'r')
zip_ref.extractall('/tmp/rps-train')
zip_ref.close()


zip_ref = zipfile.ZipFile('/content/rps-test-set.zip', 'r')
zip_ref.extractall('/tmp/rps-test')
zip_ref.close()

```

```python
import os

base_dir = '/tmp/rps-train/rps'

train_paper_dir = os.path.join(base_dir, 'paper')
train_rock_dir = os.path.join(base_dir, 'rock')
train_scissors_dir = os.path.join(base_dir, 'scissors')

print('total number of training images of paper', len(train_paper_dir))
print('total number of training images of rock', len(train_rock_dir))
print('total number of training images of scissors', len(train_scissors_dir))

```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for files in os.listdir(train_paper_dir)[:5]:
  img = mpimg.imread(os.path.join(train_paper_dir, files))
  plt.imshow(img)
  plt.show()
```
